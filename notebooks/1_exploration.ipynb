### verify the images and masks

```python
import os

from pathlib import Path

train_image_dir = "../data/ISIC_Training_Input"
mask_dir = "../data/ISIC_Training_GroundTruth"

train_images = sorted(os.listdir(train_image_dir))
mask = sorted(os.listdir(mask_dir))

print("Number of images:", len(train_images))
print("Number of mask:", len(mask))
```

```python
import cv2
import matplotlib.pyplot as plt


# Load image
image = cv2.imread("../data/ISIC_Training_Input/ISIC_0000000.jpg")

# Load mask
mask = cv2.imread("../data/ISIC_Training_GroundTruth/ISIC_0000000_segmentation.png")

print("Image shape:", image.shape)
print("Mask shape:", mask.shape)
print("Mask unique values:", set(mask.flatten()))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(mask, cmap="gray")
plt.title("Mask")
plt.axis("off")

plt.show()
```

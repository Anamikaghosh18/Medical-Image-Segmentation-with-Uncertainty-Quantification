```python
import cv2
import numpy as np
import torch
import matplotlib.pyplot as plt

image_path = "../data/ISIC_Training_Input/ISIC_0000000.jpg"
mask_path = "../data\ISIC_Training_GroundTruth\ISIC_0000000_segmentation.png"

# Load image
image = cv2.imread(image_path)

# Load mask
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

# Resize
image = cv2.resize(image, (256, 256))
mask = cv2.resize(mask, (256, 256), interpolation=cv2.INTER_NEAREST)

# Normalize
image = image.astype(np.float32) / 255.0

# Binarize mask
mask = (mask > 127).astype(np.float32)

# Convert to tensors
image_tensor = torch.from_numpy(image).permute(2, 0, 1)
mask_tensor = torch.from_numpy(mask).unsqueeze(0)

print("Image: " , image_tensor.shape)
print("Mask: " , mask_tensor.shape)
print("Mak tensor: " , mask_tensor.unique())
```
